```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
# URL of the web page you want to scrape
url = 'https://www.tfrrs.org/all_performances/IN_college_m_Indiana_IN.html?list_hnd=4153&season_hnd=608.com'

# Make an HTTP request to the web page
page = requests.get(url)
page = page.content
soup = BeautifulSoup(page,'html.parser')

```

```python
hundred_meter_table = soup.find_all('tbody',id='body',class_='body')[0]
```

```python
tbody = soup.find_all('tr', class_='allRows')
data={}
id = [1, 2, 4, 9, 10,11]
for i in id:
  data[soup.find('th',id=str(i)).text]=[]
```

```python
for i in range(22):
  try:
    data['Athlete'].append(hundred_meter_table.find_all(class_="tablesaw-priority-1")[i].text.strip())
  except Exception:
    data['Athlete'].append(None)
  try:
    data['Year'].append(hundred_meter_table.find_all(class_="tablesaw-priority-4")[i].text.strip())
  except Exception:
    data['Year'].append(None)
  try:
    data['Meet'].append(hundred_meter_table.find_all(class_="tablesaw-priority-3")[i].text.strip())
  except Exception:
    data['Meet'].append(None)
  try:
    data['Meet Date'].append(hundred_meter_table.find_all(class_="tablesaw-priority-2")[i].text.strip())
  except Exception:
    data['Meet Date'].append(None)
  try:
    data['Wind'].append(hundred_meter_table.find_all(class_="tablesaw-priority-5")[i].text.strip())
  except Exception:
    data['Wind'].append(None)
time_values = [a.text for a in hundred_meter_table.find_all('a') if re.search(r'\d+\.\d+', a.text)]
data['Time'].extend(time_values)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#Changing the First & Last Name
df_new = df['Athlete'].str.split(',', expand=True)
df_new.columns = ["Last","First"]
df_full_name = df_new['First']+ " "+ df_new['Last']
df['Athlete'] = df_full_name
```

```python
#Changing the format of Year
df_year =df['Year'].str.split("-", expand= True)
df_year.columns=["char","numeric"]
df['Year']=df_year['numeric']
```

```python
#Converting the Meet Date into epochs date
df['Meet Date']= pd.to_datetime(df['Meet Date'])
df['Meet Date'] = df['Meet Date'].astype('int64') // 10**9
```

```python
df.to_csv("/content/drive/MyDrive/Question_1.csv")
```

```python
file_path = "/content/drive/MyDrive/example_invoice.txt"

with open(file_path, 'r') as file:
    text_data = file.read()
text_data = text_data.split("\n")
text_data[0].strip()
```

```python
data_corrected=[]
pattern = re.compile(r"(\d+)([A-Za-z]+)\s+([A-Za-z]+)(GS\d{4}|GO\d{4}|GP\d{4}).*?(SHIRT WORK LS|SHIRT WORK SS|COVERALL GAS CO COTTON|PANT WORK TWILL).*?(\d+\.\d+)\s*$")
for i in range(len(text_data)):
  match = pattern.search(str(text_data[i]))
  if match:
    data_corrected.append({
            'wearer_number': match.group(1),
            'wearer_first': match.group(2),
            'wearer_last': match.group(3),
            'item_code': match.group(4),
            'item_description': match.group(5).strip(),
            'cost': match.group(6)
        })
df_corrected = pd.DataFrame(data_corrected)
```

```python
df_corrected.to_csv("/content/drive/MyDrive/Question_2.csv")
```
